```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import nltk
from nltk.corpus import stopwords
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('job_postings.csv')
```

```python
df
```

```python
user1 = 'software developer python java'
user2 = 'product manager agile'
```

```python
df['title'].fillna('', inplace=True)
df['skills_desc'].fillna('', inplace=True)
```

```python
df['title'].fillna('missing', inplace=True)
```

```python
tfidf = TfidfVectorizer()
job_vectors = tfidf.fit_transform(df['title'] + ' ' + df['skills_desc'])

user1_vector = tfidf.transform([user1])
user2_vector = tfidf.transform([user2])
```

```python
# Compute similarity
user1_sim = cosine_similarity(user1_vector, job_vectors).flatten()
user2_sim = cosine_similarity(user2_vector, job_vectors).flatten()
```

```python
user1_recommendations = df.iloc[user1_sim.argsort()[-5:]][['title']]
user2_recommendations = df.iloc[user2_sim.argsort()[-5:]][['title']]
```

```python
print(user1_recommendations)
print(user2_recommendations)
```
